```python
ftp_species = []
for spec in df_ftp.GenusSpecies.unique().tolist():
    parts = str(spec).split(' - ')
    if len(parts) > 1:
        standard = parts[1].lower()
        ftp_species.append(standard)
        
stc_species = [str(x).lower() for x in df_stc.spc_common.unique().tolist()]

matches = 0
unmatched = []
for spec in ftp_species:
    if spec not in stc_species:
        unmatched.append(spec)
    else:
        matches += 1

print("%d without match, %d matches" % (len(unmatched), matches))
print(unmatched)
```

```python
import re

def remove_symbols(s):
    return re.sub(r'[^a-zA-Z0-9 ]', '', s)

def to_camel_case(s):
    parts = s.split(' ')
    return parts[0].lower() + ''.join(word.title() for word in parts[1:])

def remove_symbols_and_camel_case(strings):
    return [to_camel_case(remove_symbols(s)) for s in strings]
```

```python
cc = remove_symbols_and_camel_case(ftp_species)
for value in ftp_species:
    print("case %s = \"%s\"" % (to_camel_case(remove_symbols(value)), value))
```
